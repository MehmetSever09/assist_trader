# Merging inflation rate on bitcoin

```python
import pandas as pd
```

```python
data_bit = pd.read_csv("../raw_data/BTCUSDT_1h.csv")
data_bit.tail()
```

```python
data_inf = pd.read_csv("../raw_data/historical_country_United_States_indicator_Interest_Rate.csv")
data_inf.tail()
```

```python
data_bit['timestamp'] = pd.to_datetime(data_bit['timestamp'])
data_bit.info()
```

```python
data_bit['datetime_M'] = data_bit['timestamp'].dt.to_period('M').dt.strftime('%Y-%m')
data_bit.head()
```

```python
data_inf['DateTime'] = pd.to_datetime(data_inf['DateTime'])
data_inf.info()
```

```python
data_inf['datetime_M'] = data_inf['DateTime'].dt.to_period('M').dt.strftime('%Y-%m')
data_inf.head()
```

```python
merged_df_inf = pd.merge(data_bit, data_inf, on=['datetime_M', 'datetime_M'], how='inner')
merged_df_inf = merged_df_inf.drop(["datetime_M", "Category", "Country", "DateTime", "Frequency", "HistoricalDataSymbol", "LastUpdate"], axis=1)
merged_df_inf
```

```python
merged_df_inf = merged_df_inf.rename(columns={'Value': 'Inflation_Rate'})
merged_df_inf
```

```python
merged_df_inf.to_csv("Biotcoin_df_Inflation_rate")
```
