### Momentum Indicators

```python
import ta 
import pandas as pd
import numpy as np
```

```python
data_path = "../raw_data/BTCUSDT_1h.csv"
df = pd.read_csv(data_path)
```

```python
!pwd
```

```python
df.head()
```

```python
df_close = df["close"]
df_open = df["open"]
df_high = df["high"]
df_low = df["low"]
```

```python
# AVERAGE DIRECTIONAL INDEX
df["adx"] = ta.trend.ADXIndicator(df_high, df_low, df_close, window=14).adx()
df["adx_neg"] = ta.trend.ADXIndicator(df_high, df_low, df_close, window=14).adx_neg()
df["adx_pos"] = ta.trend.ADXIndicator(df_high, df_low, df_close, window=14).adx_pos()
df
```

```python
df["aroon_down"] = ta.trend.AroonIndicator(df_close, window=14).aroon_down()
df["aroon"] = ta.trend.AroonIndicator(df_close, window=14).aroon_indicator()
df["aroon_up"] = ta.trend.AroonIndicator(df_close, window=14).aroon_up()
df
```

```python
df["cci"] = ta.trend.CCIIndicator(df_high, df_low, df_close, window=20, constant= 0.015).cci()
df
```

```python
df["dpo"] = ta.trend.DPOIndicator(df_close, window=20).dpo()
df
```

```python
df["ema_14"] = ta.trend.EMAIndicator(df_close, window=14).ema_indicator()
df["ema_20"] = ta.trend.EMAIndicator(df_close, window=20).ema_indicator()
df["ema_50"] = ta.trend.EMAIndicator(df_close, window=50).ema_indicator()
df["ema_100"] = ta.trend.EMAIndicator(df_close, window=100).ema_indicator()
df["ema_200"] = ta.trend.EMAIndicator(df_close, window=200).ema_indicator()
df
```

```python
df["kst"] = ta.trend.KSTIndicator(df_close, roc1= 10, roc2= 15, roc3= 20, roc4= 30,
                                  window1= 10, window2= 10, window3= 10, window4= 15, nsig= 9).kst()
df["kst_diff"] = ta.trend.KSTIndicator(df_close, roc1= 10, roc2= 15, roc3= 20, roc4= 30,
                                  window1= 10, window2= 10, window3= 10, window4= 15, nsig= 9).kst_diff()
df["kst_sig"] = ta.trend.KSTIndicator(df_close, roc1= 10, roc2= 15, roc3= 20, roc4= 30,
                                  window1= 10, window2= 10, window3= 10, window4= 15, nsig= 9).kst_sig()
df
```

```python
df["macd_line"] = ta.trend.MACD(df_close, window_slow= 26, window_fast= 12, window_sign= 9).macd()
df["macd_signal"] = ta.trend.MACD(df_close, window_slow= 26, window_fast= 12, window_sign= 9).macd_signal()
df
```

```python
df["mass_index"] = ta.trend.MassIndex(df_high, df_low, window_fast= 9, window_slow= 25).mass_index()
df
```

```python
df["parabolic_sar"] = ta.trend.PSARIndicator(df_high, df_low, df_close, step= 0.02, max_step= 0.2).psar()
df
```

```python
df["sma_14"]= ta.trend.SMAIndicator(df_close, window=14).sma_indicator()
df["sma_20"]= ta.trend.SMAIndicator(df_close, window=20).sma_indicator()
df["sma_50"]= ta.trend.SMAIndicator(df_close, window=50).sma_indicator()
df["sma_100"]= ta.trend.SMAIndicator(df_close, window=100).sma_indicator()
df["sma_200"]= ta.trend.SMAIndicator(df_close, window=200).sma_indicator()
df
```

```python
df["stc"]= ta.trend.STCIndicator(df_close, window_slow= 50, window_fast= 23, cycle= 10, smooth1= 3, smooth2= 3).stc()
df
```

```python
df["trix"]= ta.trend.TRIXIndicator(df_close, window= 15).trix()
df
```

```python
df["vortex_ind_diff"] = ta.trend.VortexIndicator(df_high, df_low, df_close, window= 14).vortex_indicator_diff()
df["vortex_neg"] = ta.trend.VortexIndicator(df_high, df_low, df_close, window= 14).vortex_indicator_neg()
df["vortex_pos"] = ta.trend.VortexIndicator(df_high, df_low, df_close, window= 14).vortex_indicator_pos()
df
```

```python
df["wma_9"] = ta.trend.WMAIndicator(df_close, window=14).wma()
df["wma_20"] = ta.trend.WMAIndicator(df_close, window=20).wma()
df["wma_50"] = ta.trend.WMAIndicator(df_close, window=50).wma()
df["wma_100"] = ta.trend.WMAIndicator(df_close, window=100).wma()
df
```

```python
df.to_csv("trend_ind_btc1h.csv")
```
