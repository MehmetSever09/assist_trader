```python
data = pd.read_csv('../../data/raw_data/BTCUSDT_daily.csv')
data["timestamp"] = pd.to_datetime(data["timestamp"])
data["timestamp"] = data["timestamp"] - pd.Timedelta(hours=3)
data["timestamp"] = data["timestamp"].dt.floor("D")
data.head()
```

```python
fig = go.Figure(data=[go.Candlestick(x=data['timestamp'],
                open=data['open'],
                high=data['high'],
                low=data['low'],
                close=data['close'])])


# fig.update_yaxes()
fig.update_layout(
    title="Bitcoin - USDT",
    yaxis_title="Close Price (USDT)",
    xaxis_title="Time",
    width=1000,
    height=1000
)

fig.show()
```
