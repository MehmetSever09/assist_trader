```python
df_1w = pd.read_csv("../../assets/data/BTC-USD_1W.csv")
df_1w = df_1w.drop(columns=["Adj Close", "Volume"])
df_1w["Date"] = pd.to_datetime(df_1w["Date"])
df_1w = df_1w[df_1w["Date"] > "2022-12-25"]
# df_1w

df_1day = pd.read_csv('../../assets/data/BTCUSDT_daily.csv')
df_1day["timestamp"] = pd.to_datetime(df_1day["timestamp"])
df_1day["timestamp"] = df_1day["timestamp"] - pd.Timedelta(hours=3)
df_1day["timestamp"] = df_1day["timestamp"].dt.floor("D")
df_1day = df_1day[df_1day["timestamp"] > "2023-01-01"]
# df_1day
```

```python
df_1w.tail()
```

```python
fig = go.Figure(
    data=[
        go.Candlestick(
            x=df_1w['Date'],
            open=df_1w['Open'],
            high=df_1w['High'],
            low=df_1w['Low'],
            close=df_1w['Close']
        ),
        go.Scatter(
            x=df_1day["timestamp"],
            y=(df_1day["close"] + df_1day['open']) / 2,
            mode="lines",
            marker= {"color":"blue"}
        )
    ])


# fig.update_yaxes()
fig.update_layout(
    title="Bitcoin - USDT",
    yaxis_title="Close Price (USDT)",
    xaxis_title="Time",
    width=1000,
    height=1000,
    xaxis = {
        "tickmode": "array",
        "tickvals": df_1w["Date"],
        "range": ["2023-01-01", "2023-06-01"]
    }
)



fig.show()
```
